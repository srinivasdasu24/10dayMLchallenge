```python
country_data['code'] = country_data.apply(lambda x:cc.convert(names=[x.name.replace(", The","")],to="iso2"),axis=1)
```

```python

def country_to_continent(code):
    try:
        cont_code = pc.country_alpha2_to_continent_code(code)
        continent = pc.convert_continent_code_to_continent_name(cont_code)
        return continent
    except:
        return None

country_data['continent'] =  country_data.apply(lambda x:country_to_continent(x.code),axis=1)
```

```python

continent_data = country_data.groupby("continent").sum()
continent_data
```

```python
fig = plt.figure(figsize=(12,8))
plt.bar(x=continent_data.index,height=continent_data.columns[-1],data=continent_data,)
plt.title("COVID-19: Continent v/s Total Reported Cases")
plt.xlabel("Continents")
plt.ylabel("Number of Reported Cases")
```
